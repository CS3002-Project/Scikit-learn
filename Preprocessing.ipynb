```python
##Feature selection: mean, sd, min, max 
def feature_extraction(window_size, data):

    i = 0
    window = window_size
    start = 0
    end = window
    remainder_rows = data.shape[0]%window
    feature_extracted_list = []

    while (i<data.shape[0]):
        total_list = []
        list = data[start:end]

    #     last_row = data[end-1:end]
    #     print(last_row)
    #     dance_label = data[-1:]
    #     print(dance_label)

        mean_array = list.mean(axis=0)
        mean_list = mean_array.tolist()
        dance_label = mean_list[-1:]
    #    time_list = time_list[:2]  
        mean_list = mean_list[:-1]
    #    mean_list = mean_list[2:]

        sd_array = list.std(axis=0)
        sd_list = sd_array.tolist()
        sd_list = sd_list[:-1]

        min_array = list.min(axis=0)
        min_list = min_array.tolist()
        min_list = min_list[:-1]

        max_array = list.max(axis=0)
        max_list = max_array.tolist()
        max_list = max_list[:-1]

    #    total_list.append(time_list)
        total_list.extend(mean_list)
        total_list.extend(sd_list)
        total_list.extend(min_list)
        total_list.extend(max_list)
        total_list.extend(dance_label)
        feature_extracted_list.append(total_list)

        i+=window
        start += window
        end += window

    df = pd.DataFrame(feature_extracted_list)
    return df

feature_extraction(10, data)
```
